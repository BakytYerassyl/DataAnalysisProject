```python
#importing pandas, numpy, matplotlib.pyplot, seaborn, os, statsmodels.formula.api, train_test_split 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import statsmodels.formula.api as smf 
from sklearn.model_selection import train_test_split 
```

```python
#storing the path to the dataset

file = './Team 6 - US Census Above 50k Predictor.csv'

#reading the file into Python through pandas

df = pd.read_csv(file)

df = pd.read_csv('./Team 6 - US Census Above 50k Predictor.csv',
                 na_values = ' ?') #reading the file with all the missing values replace by ' ?'

#renaming all the columns 

df.columns=['Age','Job/Occupation as classified by the census', 
            'Type of employment', 'Number of hours currently working per week', 
            'Capital gain/Capital loss', 'Highest level of education', 'Completed years of education', 
            'Marital Status', 'Relationship inside the household', 'Race', 'Gender','Country of Origin','Income']

#returning the first 10 rows for the object based on position

df.head(n=10)
```

Data cleaning and feature engineering

```python
#replacing all the ' ?' by 'nan'

df=df.replace(' ?', np.nan)

#replacing the nan variables by the mode of the column

df['Job/Occupation as classified by the census']= df['Job/Occupation as classified by the census'].replace(np.nan, df['Job/Occupation as classified by the census'].mode()[0])

#getting unique values of Series object.

df['Job/Occupation as classified by the census'].unique()

df['Type of employment']= df['Type of employment'].replace(np.nan, df['Type of employment'].mode()[0])

df['Type of employment'].unique()

df['Country of Origin']= df['Country of Origin'].replace(np.nan, df['Country of Origin'].mode()[0])

df['Type of employment'].unique()

# regrouping the variables in a new variables/bins for helping the regression

df['M_Country_of_Origin'] = ''
df['M_Country_of_Origin'] = np.where(df['Country of Origin'] == " United-States","US","Expat")

# regrouping the variables in a new variables/bins for helping the regression

df['m_Education_Level'] =''
df['m_Education_Level'] = np.where(df['Highest level of education'] == " Bachelors","Graduated", df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " Assoc-acdm","Graduated",df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " Some-college","Graduated",df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " 11th","Undergraduated",df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " 5th-6th","Undergraduated",df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " 1st-4th","Undergraduated",df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " 7th-8th","Undergraduated",df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " 9th","Undergraduated",df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " 10th","Undergraduated",df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " 12th","Undergraduated",df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " HS-grad","Graduated",df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " Masters","Advanced", df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " Doctorate","Advanced", df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " Prof-school","Advanced", df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " Preschool","Undergraduated", df['m_Education_Level'])
df['m_Education_Level'] = np.where(df['Highest level of education'] == " Assoc-voc","Undergraduated", df['m_Education_Level'])

# regrouping the variables in a new variables/bins for helping the regression

df['m_Job_occupation'] = ''
df['m_Job_occupation'] = np.where(df['Job/Occupation as classified by the census'] == " Armed-Forces "," Manual", df['m_Job_occupation'])
df['m_Job_occupation'] = np.where(df['Job/Occupation as classified by the census'] == " Protective-serv"," Manual", df["m_Job_occupation"])
df['m_Job_occupation'] = np.where(df['Job/Occupation as classified by the census'] == " Priv-house-serv"," Manual", df["m_Job_occupation"])
df['m_Job_occupation'] = np.where(df['Job/Occupation as classified by the census'] == " Exec-managerial"," Manager", df["m_Job_occupation"])
df['m_Job_occupation'] = np.where(df['Job/Occupation as classified by the census'] == " Tech-support"," Other", df["m_Job_occupation"])
df['m_Job_occupation'] = np.where(df['Job/Occupation as classified by the census'] == " Farming-fishing"," Manual", df["m_Job_occupation"])
df['m_Job_occupation'] = np.where(df['Job/Occupation as classified by the census'] == " Adm-clerical"," Other", df["m_Job_occupation"])
df['m_Job_occupation'] = np.where(df['Job/Occupation as classified by the census'] == " Sales"," Other", df["m_Job_occupation"])
df['m_Job_occupation'] = np.where(df['Job/Occupation as classified by the census'] == " Craft-repair"," Manual", df["m_Job_occupation"])
df['m_Job_occupation'] = np.where(df['Job/Occupation as classified by the census'] == " Transport-moving"," Manual", df["m_Job_occupation"])
df['m_Job_occupation'] = np.where(df['Job/Occupation as classified by the census'] == " Handlers-cleaners"," Manual", df["m_Job_occupation"])
df['m_Job_occupation'] = np.where(df['Job/Occupation as classified by the census'] == " Prof-specialty"," Manager", df["m_Job_occupation"])
df['m_Job_occupation'] = np.where(df['Job/Occupation as classified by the census'] == " Machine-op-inspct"," Manual", df["m_Job_occupation"])
df['m_Job_occupation'] = np.where(df['Job/Occupation as classified by the census'] == " Other-service"," Other", df["m_Job_occupation"])

# replace the two variables in a new bins (1 or 0) for helping the regression

df['M_income'] = np.where(df['Income'] == ' <=50K',1,0)

# regrouping the variables in a new variables/bins for helping the regression

df['m_Employment'] =''
df['m_Employment'] = np.where(df['Type of employment'] ==' Federal-gov',"Gov", df['m_Employment'])
df['m_Employment'] = np.where(df['Type of employment'] ==' State-gov',"Gov", df['m_Employment'])
df['m_Employment'] = np.where(df['Type of employment'] ==' Local-gov',"Gov", df['m_Employment'])

df['m_Employment'] = np.where(df['Type of employment'] ==' Private',"Private", df['m_Employment'])
df['m_Employment'] = np.where(df['Type of employment'] ==' Self-emp-inc',"Private", df['m_Employment'])
df['m_Employment'] = np.where(df['Type of employment'] ==' Self-emp-not-inc',"Private", df['m_Employment'])

df['m_Employment'] = np.where(df['Type of employment'] ==' Without-pay',"Other", df['m_Employment'])
df['m_Employment'] = np.where(df['Type of employment'] ==' Never-worked',"Other", df['m_Employment'])

# regrouping the variables in a new variables/bins for helping the regression

df['m_marital_stat'] = ""
df['m_marital_stat'] = np.where(df['Marital Status'] == " Married-AF-spouse"," Married", df["m_marital_stat"])
df['m_marital_stat'] = np.where(df['Marital Status'] == " Never-married"," Single", df["m_marital_stat"])
df['m_marital_stat'] = np.where(df['Marital Status'] == " Married-civ-spouse"," Married", df["m_marital_stat"])
df['m_marital_stat'] = np.where(df['Marital Status'] == " Divorced"," Single", df["m_marital_stat"])
df['m_marital_stat'] = np.where(df['Marital Status'] == " Widowed"," Single", df["m_marital_stat"])
df['m_marital_stat'] = np.where(df['Marital Status'] == " Separated"," Single", df["m_marital_stat"])
df['m_marital_stat'] = np.where(df['Marital Status'] == " Married-spouse-absent"," Married", df["m_marital_stat"])

# regrouping the variables in a new variables/bins for helping the regression

df['m_Race'] =''
df['m_Race'] = np.where(df['Race'] ==' White',"White", df['m_Race'])
df['m_Race'] = np.where(df['Race'] ==' Black',"Black", df['m_Race'])
df['m_Race'] = np.where(df['Race'] ==' Asian-Pac-Islander',"Other", df['m_Race'])
df['m_Race'] = np.where(df['Race'] ==' Amer-Indian-Eskimo',"Other", df['m_Race'])
df['m_Race'] = np.where(df['Race'] ==' Other',"Other", df['m_Race'])

#using pd.cut() for regrouping int variables in bins

bins= [1,40,45,90]
labels = ['Part Time','Work Life Balance','Overtime']
df['m_WorkHour'] = pd.cut(df['Number of hours currently working per week'], bins=bins, labels=labels, right=False)

#using pd.cut() for regrouping int variables in bins

bins= [16,29,38,49,91]
labels = ['1','2','3','4']
df['m_age'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
#converting categorical variable into dummy variables.

df_dummies = pd.get_dummies(data = df,
              columns = ['m_Race','m_marital_stat', 
                         'M_Country_of_Origin', 'm_Education_Level', 'm_Job_occupation', 'm_Employment', 'm_WorkHour'])

#Training and Testing

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

df_data = df_dummies.drop(['M_income'],
                    axis = 1)


# preparing response variable data

df_target = df_dummies.loc[ : , 'M_income']
```

```python
# preparing training and testing sets 

x_train, x_test, y_train, y_test = train_test_split(
            df_data,
            df_target,
            test_size = 0.25,
            random_state = 1223)

#converting in CSV databases and exporting the trainings and the testings 

x_train.to_csv(r'./data/train_features.csv',index = True)
x_test.to_csv(r'./data/test_features.csv', index = True)
y_train.to_csv(r'./data/train_y.csv', index = True)
y_test.to_csv(r'./data/test_y.csv', index = True)

#returning  all the x_train 

x_train.head()
```
